```python
# while True:
df = get_votes_per_state()
df = df[['name', 'code', 'JAIR BOLSONARO', 'LULA']]
df[['JAIR BOLSONARO', 'LULA']] = (df[['JAIR BOLSONARO', 'LULA']].astype(int)).astype(int)
df.columns = ['name', 'code', 'jair', 'lula']

df['now_total'] = df[['jair', 'lula']].sum(axis=1)
comp = pd.concat([primeiro_turno, df], axis=1)
comp['proporcao_urnas'] = comp['now_total']/comp['votos_validos']

comp['cenario_atual'] = comp['lula']/comp['now_total']
comp['jair_projecao'] = comp['jair'] * comp['votos_validos']/comp['now_total']
comp['lula_projecao'] = comp['lula'] * comp['votos_validos']/comp['now_total']
comp['lula_prop'] = comp['lula']/comp['now_total']

# otimista
comp['lula_projecao_otimista'] = (comp['lula_prop']*comp['proporcao_urnas'] + 
                                  comp['cenario_favoravel_a_lula']*(1 - comp['proporcao_urnas'])
                                 ) * comp['votos_validos']

comp['lula_projecao_pessimista'] = (comp['lula_prop']*comp['proporcao_urnas'] + 
                                  comp['cenario_favoravel_a_bolsonaro']*(1 - comp['proporcao_urnas'])
                                 ) * comp['votos_validos']

comp['lula_projecao_neutra'] = (comp['lula_prop']*comp['proporcao_urnas'] + 
                                  comp['cenario_intermediario']*(1 - comp['proporcao_urnas'])
                                 ) * comp['votos_validos']


comp['lula_projecao_ruidosa'] = (comp['lula_prop']
                                 ) * comp['votos_validos']


now = {
    'Pessimista': comp['lula_projecao_pessimista'].sum()/comp['votos_validos'].sum(),
    'Neutra': comp['lula_projecao_neutra'].sum()/comp['votos_validos'].sum(),
    'Otimista': comp['lula_projecao_otimista'].sum()/comp['votos_validos'].sum(),
    'Simples': comp['lula_projecao_ruidosa'].sum()/comp['votos_validos'].sum(),
    'hora': pd.to_datetime(dt.now())
      }
resultados.append(now)

hora = str(pd.to_datetime(dt.now()))
comp.to_csv(f"../data/dados_{hora}")

resultados_para_plot = pd.DataFrame(resultados).melt(id_vars='hora')
resultados_para_plot['Minuto'] = resultados_para_plot.hora.apply(lambda h: pd.to_datetime(h).strftime("%H:%M"))

#     plt.figure(dpi=150)
#     sns.lineplot(data = resultados_para_plot, x='hora', y='value', hue='variable', marker='o')
#     plt.ylim(.4,0.6)
#     plt.hlines(0.5, resultados_para_plot.hora.min(), hora_final)
#     plt.legend(frameon=False, title='')
#     plt.ylabel("Projeção de % para Lula")
#     plt.xticks(rotation = 45);

#     print(hora)
#     plt.show()

#     time.sleep(180)
```
