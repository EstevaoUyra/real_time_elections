```python
states = pd.read_csv("../data/estados_demografia.csv")
```

```python
votes = states.code.str.lower().apply(summarize_state_res)
primeiro_turno = pd.concat([states, votes], axis=1)
```

```python
candidatos_1o_turno = ['JAIR BOLSONARO', 'LULA', 'SIMONE TEBET', 'CIRO GOMES',
       'SORAYA THRONICKE', 'FELIPE D&apos;AVILA', 'PADRE KELMON',
       'SOFIA MANZANO', 'VERA', 'LÉO PÉRICLES', 'CONSTITUINTE EYMAEL']
primeiro_turno[candidatos_1o_turno] = primeiro_turno[candidatos_1o_turno].astype(int)
primeiro_turno['votos_validos'] = primeiro_turno[candidatos_1o_turno].sum(axis=1)
primeiro_turno[['abstencoes', 'brancos', 'nulos']] = primeiro_turno[['abstencoes', 'brancos', 'nulos']].astype(int)
primeiro_turno['total_de_votos'] = primeiro_turno[['brancos', 'nulos', 'votos_validos']].sum(axis=1)
```

```python
coluna_de_totais = 'votos_validos'

primeiro_turno[['perc_bolsonaro', 'perc_lula']] = (
    (primeiro_turno[['JAIR BOLSONARO', 'LULA']].T / primeiro_turno[coluna_de_totais]).T
)
primeiro_turno['cenario_favoravel_a_lula'] = 1 - primeiro_turno['perc_bolsonaro']
primeiro_turno['cenario_favoravel_a_bolsonaro'] = primeiro_turno['perc_lula']
primeiro_turno['cenario_intermediario'] = primeiro_turno['perc_lula']/(primeiro_turno['perc_bolsonaro'] + primeiro_turno['perc_lula'])

primeiro_turno
```

```python
primeiro_turno.to_csv("../data/resultados_1o_turno_fim.csv", index=False)
```
